```python
# Importing Libraries
from bokeh.plotting import figure
from bokeh.io import output_file, show

# Defining values
x = [3, 7.5, 10]
y = [3, 6, 9]

# Creating the file
output_file("Triangle.html")

# Using the figure object
f = figure()

# Plotting
f.triangle(x, y)

# Getting an output
show(f)
```

```python
# Import bokeh, pandas
import pandas as pd
from bokeh.plotting import figure
from bokeh.io import output_file, show

# Import the csv file
women_bachelors = pd.read_csv("bachelors.csv")

# Defining x and y values
x = women_bachelors["Year"]
y = women_bachelors["Engineering"]

# Creating an output file
output_file("Women_bachelor_percentage.html")

# Creating a figure object
f = figure()

# Creating the plot using the figure object
f.line(x, y)

# Showing the plot
show(f)
```

```python
import pandas
from bokeh.plotting import figure, output_file, show
 
p=figure(plot_width=500,plot_height=400, tools='pan')
 
p.title.text="Cool Data"
p.title.text_color="Gray"
p.title.text_font="times"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Date"
p.yaxis.axis_label="Intensity"    
 
p.line([1,2,3],[4,5,6])
output_file("graph.html")
show(p)
```

```python
# Importing pandas and bokeh
import pandas as pd
from bokeh.plotting import figure
from bokeh.io import output_file, show

# Importing the files
weather_data = pd.read_excel("verlegenhuken.xlsx")

# Choosing x and y values
x = weather_data["Temperature"]/10
y = weather_data["Pressure"]/10

# Prepare the output file
output_file("weather_data.html")

# Create the figure object
f = figure(plot_width = 700, plot_height = 600, tools = 'pan')

f.title.text = 'Temperature and Air Pressure'
f.title.text_font = 'roboto'
f.title.text_color = 'Gray'
f.title.text_font_style = 'bold'
f.xaxis.axis_label = 'Temperature(°C)'
f.yaxis.axis_label = 'Pressure(hPa)'

# Plotting
f.circle_dot(x, y, size = 0.5)

# Show
show(f)
```

```python
import pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show

df = pandas.read_csv("adbe.csv", parse_dates = ['Date'])

p = figure(width = 500, height = 250, x_axis_type = "datetime")
p.line(df["Date"], df["Close"], color = 'Orange', alpha = 0.5)
output_file("adbe.html")

show(p)
```
